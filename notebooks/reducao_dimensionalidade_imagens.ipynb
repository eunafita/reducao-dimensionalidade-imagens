Importando as bibliotecas necessárias para processamento e visualização de imagens:

```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
```

Funções para converter a imagem de colorida para escala de cinza e binarizar a imagem:

```python
# Função para converter para escala de cinza
def rgb_to_gray(image):
    # A fórmula de conversão é baseada em médias ponderadas para simular a percepção humana de cores
    r, g, b = image[:,:,0], image[:,:,1], image[:,:,2]
    gray_image = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray_image

# Função para binarizar a imagem
def binarize_image(gray_image, threshold=127):
    binary_image = np.where(gray_image > threshold, 255, 0)  # 255 para branco, 0 para preto
    return binary_image
```

Carregar a imagem selecionada e converter para escala de cinza e binarizar a imagem:

```python
# Carregar a imagem colorida
img = imread('./images/cachorro_coelho.jpeg')

# Converter para escala de cinza
gray_img = rgb_to_gray(img)

# Binarizar a imagem
binary_img = binarize_image(gray_img)
```

Mostrar as imagens:

```python
# Mostrar as imagens
plt.figure(figsize=(10,10))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Imagem em Escala de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_img, cmap='gray')
plt.title('Imagem Binarizada')
plt.axis('off')

plt.show()
```
